```
import tensorflow_datasets as tfds
imdb,info= tfds.load("imdb_reviews",with_info=True,as_supervised=True)
train_data,test_data = imdb['train'], imdb['test']
training_sentences=[]
training_labels=[]
testing_sentences=[]
testing_labels=[]
for s,l in train_data:
  training_sentences.append(str(s.numpy()))
  training_labels.append(l)
for s,l in test_data:
  testing_sentences.append(str(s.numpy()))
  testing_labels.append(l)
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
tokenizer=Tokenizer(10000,oov_token="<OOV>")
tokenizer.fit_on_texts(training_sentences)
word_index=tokenizer.word_index
sequences=tokenizer.texts_to_sequences(training_sentences)
padded=pad_sequences(sequences,maxlen=1200,truncating='post')
testing_sequences=tokenizer.texts_to_sequences(testing_sentences)
testing_padded=pad_sequences(testing_sequences,maxlen=1200)
```

```
import tensorflow as tf
import numpy as np
model=tf.keras.models.Sequential([
                                 tf.keras.layers.Embedding(10000,16,input_length=1200),
                                 tf.keras.layers.Flatten(),
                                 tf.keras.layers.Dense(6,activation='relu'),
                                 tf.keras.layers.Dense(1,activation='sigmoid') 
])
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
model.fit(padded,np.array(training_labels),epochs=10,validation_data=(testing_padded,np.array(testing_labels)))
new_sents = [
    'This movie was awesome. Really loved it.',
    'I almost fell asleep watching this.',
    'This movie is so stupid.',
    'I wouldn\'t watch this movie unless there\'s a gun pointed at me.',
    'This movie is so hilarious. Much love to the director!'
]
pred_sequences=tokenizer.texts_to_sequences(new_sents)
pred_padded=pad_sequences(pred_sequences,maxlen=1200)
output=model.predict(pred_padded)
for i in range(0,len(new_sents)):
    print('Review:'+new_sents[i]+' '+'   Sentiment->'+str(output[i])+'\n')
```
